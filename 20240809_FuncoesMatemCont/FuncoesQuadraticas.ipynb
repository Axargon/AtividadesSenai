```python
import matplotlib.pyplot as plt
import numpy as np

a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))

delta = b**2-(4*a*c)

if delta >0 :
  rz1 = (-b - (delta)**(1/2))/2*a
  rz2 = (-b + (delta)**(1/2))/2*a
  print(f"A equação tem 2 raízes reais: x1= {rz1}, x2= {rz2}")
elif (delta == 0 ):
  print(f"A equação tem 1 raíz real: x1 = {-b/(2*a)}")
else:
  print("A equação não tem raízes reais")

r1=(-b/(2*a))

x=np.arange(r1-50,r1+50,1,dtype= int)
y= a*x**2+b*x + c

plt.plot(x,y)
plt.xlabel('x',fontsize=15)
plt.ylabel('y',fontsize = 15)
plt.title('Função quadrática')
plt.legend(f"f(x)={a}x^2+{b}x+{c}")
plt.show()
```
