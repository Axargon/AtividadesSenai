```python
import numpy as np
#variaveis x
x=np.array([-2,-1,0,1,2])

#lei de formacao
y=2*x

print(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(x,y)
plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize = 15)
plt.title('Função Linear')
plt.legend('y=2x')
plt.show()
```

Um usuario fornecera dois pontos pertencentes a uma reta, crie o codigo que exibira a lei de formacao da reta e exibira o grafico desta reta

```python
print("Informe o primeiro ponto")
x1 = float(input("x1: "))
y1 = float(input("y1: "))
print("Informe o segunto ponto")
x2 = float(input("x2: "))
y2 = float(input("y2: "))

a = (y2-y1)/(x2-x1)
b = y1-a*x1

print(f"A equação desta reta é:f(x)={a}x+{b}")

x=np.arange(-10,10,1,dtype =int)
y=a*x+b
plt.plot(x,y)
plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize = 15)
plt.title('Função Linear')
plt.legend(f'{a}x+{b}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1:
#plt.figure().set_figheight(2,3)
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('Serra')
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('Linear')
plt.plot(x,y)
plt.tight_layout(pad=5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1:
#plt.figure().set_figheight(1,2)
x=np.array([0,1,2,3])
y=2*x+6
plt.subplot(2,2,1)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('2x+6')
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=3*x+6
plt.subplot(2,2,2)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('3x+6')
plt.plot(x,y)

#plot3:
x=np.array([0,1,2,3])
y=x-(1/5)
plt.subplot(2,2,3)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('x-(1/5)')
plt.plot(x,y)

#plot4:
x=np.array([0,1,2,3])
y=-6*x-2
plt.subplot(2,2,4)
plt.xlabel('x',fontsize =14)
plt.ylabel('y',fontsize = 15)
plt.title('-6x-2')
plt.plot(x,y)
plt.tight_layout(pad=1)
```
