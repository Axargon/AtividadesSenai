### 1.Carregando os dados

```python
import pandas as pd
df_2000 =pd.read_csv("dados_gasolina2000.csv")
df_2010 = pd.read_csv("dados_gasolina_2010.csv")
```

### 2 Analisando a estrutura dos dados

```python
df_2000.rename(columns={"DT.Inicial": "DATA INICIAL"
                        ,"DT.Final":"DATA FINAL"
                        ,"Divisão territorial":"REGIÃO"
                        ,"Item":"PRODUTO"}, inplace=True)
df_2000.info()
```

```python
df_2010.info()
```

### 3 Carregue os conjuntos de dados "dados_gasolina2000.csv" e "dados_gasolina_2010.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.


```python
df= pd.concat([df_2000,df_2010])
df.head()
```

### 4 Investigue as colunas e entenda o conjunto de dados usando o head() e info()


```python
df.info()
```

### 5 Selecione a terceira entrada da coluna DATA INICIAL e verifique seu tipo.


```python
df["DATA INICIAL"].head(3)
#df.head(3)
```

### 6 Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
colunasData = ["DATA INICIAL","DATA FINAL"]
df[colunasData] = df[colunasData].apply(pd.to_datetime, errors = "coerce")
# errors = "coerce", os valores náo numericos nas colunas especificadas serao substituidos por NAN
df.info()
```

### 7 Crie uma nova coluna para representar o ano e o mês(mm-aaaa), utilizando a coluna DATA FINAL como referência.

```python
df['MES/ANO']= df['DATA FINAL'].apply(lambda valor: str(valor.month).zfill(2) + '/' +str(valor.year))
```

### 8 Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
df['PRODUTO'].value_counts()
```

### 9 Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_filtro = df['PRODUTO'] == 'GASOLINA COMUM'
df3 = df[df_filtro]

df3.head()
```

### 10 Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
media_preco_revenda = df3[df3['MES/ANO']  == '08/2008']['PREÇO MÉDIO REVENDA'].mean()

media_preco_revenda
```

### 11 Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
media_SP = df3[(df3['MES/ANO']== '05/2014')&(df3['ESTADO']== 'SAO PAULO')]['PREÇO MÉDIO REVENDA'].mean()
media_SP
```
